```python
import os
import sys
from pathlib import Path
from dotenv import load_dotenv

# Add src directory to Python path so imports work
src_path = Path.cwd() / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))
    print(f"Added to sys.path: {src_path}")

load_dotenv()

# Create an instance
from main import AlzKBPipeline
pipeline = AlzKBPipeline(base_dir='test/')
raw_dir = str(getattr(pipeline, 'raw_dir'))
```

# AOP-DB

```python
from parsers import AOPDBParser

# Initialize parser
aopdb_parser = AOPDBParser(
    data_dir=raw_dir,
    mysql_config={
        'host': 'localhost',
        'user': os.getenv('MYSQL_USERNAME'),
        'password': os.getenv('MYSQL_PASSWORD'),
        'database': os.getenv('MYSQL_DB_NAME', 'aopdb')
    }
)

# Download and parse
aopdb_parser.download_data()
aopdb_data = aopdb_parser.parse_data()
parsed_data = {}
parsed_data['aopdb'] = aopdb_data

# Export to TSV
pipeline.export_to_tsv(parsed_data)
```
