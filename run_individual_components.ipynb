# AlzKB — Run Individual Parsers

Use this notebook to download and parse data from individual sources.
Run the setup cell first, then run any source cell independently.

For the full pipeline, use the CLI instead:
```bash
python src/main.py
```

````python
import os
import sys
from pathlib import Path
from dotenv import load_dotenv

# Add src to path
src_path = Path.cwd() / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

load_dotenv()

base_dir = Path.cwd()
raw_dir = str(base_dir / "data" / "raw")
processed_dir = base_dir / "data" / "processed"
Path(raw_dir).mkdir(parents=True, exist_ok=True)
processed_dir.mkdir(parents=True, exist_ok=True)

# Load disease scope from config
from main import load_config, export_tsv
project_config, databases, ontology_mappings = load_config()
disease_scope = project_config.get("disease_scope", {})

print(f"Disease scope: {list(disease_scope.get('primary_terms', []))}")
print(f"Raw data directory: {raw_dir}")
````

## AOP-DB

````python
from parsers import AOPDBParser

parser = AOPDBParser(
    data_dir=raw_dir,
    mysql_config={
        'host': 'localhost',
        'user': os.getenv('MYSQL_USERNAME'),
        'password': os.getenv('MYSQL_PASSWORD'),
        'database': os.getenv('MYSQL_DB_NAME', 'aopdb')
    }
)
parser.download_data()
data = parser.parse_data()
export_tsv({'aopdb': data}, processed_dir)
````

## DisGeNET

````python
from parsers import DisGeNETParser

parser = DisGeNETParser(
    data_dir=raw_dir,
    api_key=os.getenv('DISGENET_API_KEY'),
    disease_scope=disease_scope,
)
parser.download_data()
data = parser.parse_data()
export_tsv({'disgenet': data}, processed_dir)
````

## DrugBank

````python
from parsers import DrugBankParser

parser = DrugBankParser(
    data_dir=raw_dir,
    username=os.getenv('DRUGBANK_USERNAME'),
    password=os.getenv('DRUGBANK_PASSWORD'),
)
parser.download_data()
data = parser.parse_data()
export_tsv({'drugbank': data}, processed_dir)
````
